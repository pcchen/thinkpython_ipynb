# IPython Notebook Interact

First import the `interact` module

```python
from IPython.html.widgets import interact
```

Define some function

```python
def f(x):
    print(x)
```

Start to **interact!**

```python
interact(f, x=3)
```

Set the range of the variable

```python
interact(f, x=(1,15))
```

Set the increment.

```python
interact(f, x=(2,3,0.1))
```

What if we use a string?

```python
interact(f, x="Hello")
```

How about a list of string?

```python
interact(f, x=["Hello"])
```

```python
interact(f, x=["Hello","Goodmorning", "Goodnight"])
```

Try a dictionary?

```python
interact(f, x={"Hello":1,"Goodmorning":2, "Goodnight":3})
```

Interactive plot?

First define your plot function with some parameters that can be tuned.

```python
%pylab inline
```

```python
x = linspace(-10, 10, 300)
a = 3
plot(x, sin(x**2 + a)+cos(x))
```

```python
def draw(a):
    x = linspace(-10, 10, 300)
    plot(x, sin(x**2 + a)+cos(x))
```

```python
draw(3)
```

```python
interact(draw, a=(-10.,10.))
```

You can tune multiple parameters too.

```python
def draw2(a, b):
    x = linspace(-10, 10, 300)
    plot(x, sin(x**2 + a)+cos(x+b))
```

```python
interact(draw2, a=(-10.,10.), b=(-10., 10.))
```
